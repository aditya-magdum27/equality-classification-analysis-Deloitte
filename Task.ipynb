```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("D:\\Python\\Task 5 Equality Table.xlsx")
```

```python
def classify(score):
    if -10 <= score <= 10:
        return "Fair"
    elif -20 <= score < -10 or 10 < score <= 20:
        return "Unfair"
    else:
        return "Highly Discriminative"
```

```python
df["Equality class"] = df["Equality Score"].apply(classify)
```

```python
df.to_excel("Equality_Table_Updated.xlsx", index=False)
```

```python
import matplotlib.pyplot as plt

df["Equality class"].value_counts().plot(kind="bar")

plt.xlabel("Equality Class")
plt.ylabel("Count")
plt.title("Equality Distribution")
plt.show()
```
